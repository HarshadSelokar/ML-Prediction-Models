```python
import pandas as pd

data = pd.read_csv("filename.csv")

data.head()
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

print("Model initialized:", model)

model.fit(data[["area"]], data["price"])  #training data

model.predict([[2500]]) #testing data
```

```python
model.coef_ * 5000 + model.intercept_
```

```python
import matplotlib.pyplot as plt

plt.scatter(data["area"], data["price"], color='green')
plt.title("House Price prediction")
```

```python
plt.plot(data["area"], model.predict(data[["area"]]), color='green')
plt.xlabel("Area in sq.ft")
```
